2. urllib.request 모듈

  - 내장 모듈이므로 별도로 설치할 필요 없음

```python
import urllib.request

# 웹 페이지 접근
res = urllib.request.urlopen("https://news.v.daum.net/v/20210904181034626") #메서드 호출 후 객체 res에 저장
print(res)


# response한 내용 읽기 
#html 객체
html = res.read() 
```

- - -
**3. BeutifulSoup**

 * html 변수 내용은 문자 -> 실제 html로 구조화하여 파싱해야 할 때 사용
 
 * html = 문서 / soup = html이 활성화된 파싱된 내용

```python
#beautifulsoup import
from bs4 import BeautifulSoup

# 'html.parser'를 활용하여 활성화된 html 객체 파싱함
soup = BeautifulSoup(html, 'html.parser') 
```

- 예제 1. 뉴스 기사 제목 내용 크롤링하기



```python
# select_one('CSS선택자') 사용


# class 속성 값을 선택자로 사용
tag = soup.select_one('h3.tit_view') 
print(tag)

# tag에서 text만 뽑아오고 싶다면?
# text, get_text() 사용하기
tag.get_text()
```

- - -
- 예제 2. 뉴스 기사 내용 메인 이미지 크롤링하기



```python
tag2 = soup.select_one('img.thumb_g_article')
print(tag2)
```

```python
# tag객체['속성']으로 속성의 값을 반환한다.
src = tag2["src"]
print(src)
```

```python
# urllib.request.urlretrieve(이미지 src 변수명, 저장할 위치)

urllib.request.urlretrieve(src, 'image1.jpg')
```
